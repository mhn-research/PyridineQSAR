Generate molecular information from SMILES and Display 2D Structure

```python
!pip install rdkit
!pip install pubchempy

from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import rdMolDescriptors
import pubchempy as pcp

def get_molecule_info(smiles):
    # Validate SMILES input
    if not isinstance(smiles, str) or not smiles.strip():
        raise ValueError("SMILES must be a non-empty string")

    # Convert SMILES to molecule object
    mol = Chem.MolFromSmiles(smiles)
    if mol is None:
        raise ValueError(f"Invalid SMILES string: {smiles}")

    # Get molecular information
    info = {
        'smiles': smiles,
        'molecule': mol,
        'formula': rdMolDescriptors.CalcMolFormula(mol),
        'num_atoms': mol.GetNumAtoms(),
        'num_bonds': mol.GetNumBonds()
    }

    # Get IUPAC name using PubChemPy
    try:
        compounds = pcp.get_compounds(smiles, 'smiles')
        if compounds:
            info['iupac_name'] = compounds[0].iupac_name if compounds[0].iupac_name else "Name not available"
        else:
            info['iupac_name'] = "Compound not found in PubChem"
    except Exception as e:
        info['iupac_name'] = f"Error retrieving name: {str(e)}"

    return info

def display_molecule_image(mol, size=(300, 300)):
    try:
        img = Draw.MolToImage(mol, size=size)
        img.show()
        return img
    except Exception as e:
        print(f"Error generating molecule image: {e}")
        return None

def main():
    # Enter SMILES descriptor
    smiles = 'c1(cc(cc(n1)c1ccccn1)c1ccccc1)c1ncccc1'

    try:
        # Get molecular information
        mol_info = get_molecule_info(smiles)

        # Print properties
        print("=" * 50)
        print("MOLECULAR INFORMATION")
        print("=" * 50)
        print(f"SMILES:          {mol_info['smiles']}")
        print(f"Chemical Name:   {mol_info['iupac_name']}")
        print(f"Molecular Formula: {mol_info['formula']}")
        print(f"Number of Atoms:   {mol_info['num_atoms']}")
        print(f"Number of Bonds:   {mol_info['num_bonds']}")
        print("=" * 50)

        # Show 2D diagram
        print("Generating molecular structure image...")
        display_molecule_image(mol_info['molecule'])

    except ValueError as ve:
        print(f"Input Error: {ve}")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")

if __name__ == "__main__":
    main()
```
